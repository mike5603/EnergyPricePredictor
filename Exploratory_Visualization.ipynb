```python
import pandas as pd
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.palettes import viridis
import math
output_notebook()
```

```python
RealTime = pd.read_pickle('realtime_combined.pkl')
DayAhead = pd.read_pickle('dayahead_combined.pkl')
```

```python
RealTime_NP15 = RealTime.loc[RealTime['region']=='NP15']
DayAhead_NP15 = DayAhead.loc[DayAhead['region']=='NP15']
```

```python
dayOfWeek = RealTime_NP15.groupby('Day Of Week String').mean().sort_values('Day of Week')
```

```python
days = ['Monday','Tuesday', 'Wednesday','Thursday','Friday','Saturday','Sunday']
p = figure(x_range=days,title='Average Price by Day of Week')
p.vbar('Day Of Week String',top = 'total',source = dayOfWeek,width=0.9)
p.title.text_font_size='14pt'
p.yaxis.axis_label='Average Price'
p.yaxis.axis_label_text_font_size='14pt'
p.yaxis.major_label_text_font_size='12pt'
p.xaxis.axis_label='Day of Week'
p.xaxis.axis_label_text_font_size='14pt'
p.xaxis.major_label_text_font_size='12pt'
p.legend.label_text_font_size='14pt'
p.y_range.start = 25
p.xaxis.major_label_orientation = math.pi/4
show(p)
```

```python
Month = RealTime_NP15.groupby('Month Name').mean().sort_values('Month')
p = figure(x_range=list(Month.index),title='Average Price by Month')
p.vbar('Month Name',top = 'total',source = Month,width=0.9)
p.title.text_font_size='14pt'
p.yaxis.axis_label='Average Price'
p.yaxis.axis_label_text_font_size='14pt'
p.yaxis.major_label_text_font_size='12pt'
p.xaxis.axis_label='Month'
p.xaxis.axis_label_text_font_size='14pt'
p.xaxis.major_label_text_font_size='12pt'
p.legend.label_text_font_size='14pt'
p.y_range.start = 20
p.xaxis.major_label_orientation = math.pi/4
show(p)
```

```python
RealTime_NP15["Average Temp"] = RealTime_NP15[['Redding TAVG', 'Sacremento TAVG','Fresno TAVG','Bakersfield TAVG','Los Angeles TAVG']].mean(axis=1)
```

```python
from scipy.optimize import curve_fit

def quad(x,a,b,c):
    return a*x**2+b*x+c
popt, pcov = curve_fit(quad, RealTime_NP15['Average Temp'], RealTime_NP15['total'])
x_fit = np.linspace(42,92,500)
y_fit = [quad(x,*popt) for x in x_fit]
```

```python
p = figure(title='Daily Price vs Average Daily Temperature')
p.circle(x = 'Average Temp', y = 'total',source = RealTime_NP15)
p.line(x=x_fit,y=y_fit,line_width=3,line_color='red')
p.title.text_font_size='14pt'
p.yaxis.axis_label='Average Price'
p.yaxis.axis_label_text_font_size='14pt'
p.yaxis.major_label_text_font_size='12pt'
p.xaxis.axis_label='Average Temperature'
p.xaxis.axis_label_text_font_size='14pt'
p.xaxis.major_label_text_font_size='12pt'
p.legend.label_text_font_size='14pt'
#p.y_range.start = 20
p.xaxis.major_label_orientation = math.pi/4
show(p)
```

```python
RealTime_NP15["Average Wind"] = RealTime_NP15[['Redding AWND', 'Sacremento AWND','Fresno AWND','Bakersfield AWND','Los Angeles AWND']].mean(axis=1)
```

```python
def exp(x,a,b,c):
    return a * np.exp(-b * x) + c
popt, pcov = curve_fit(exp, RealTime_NP15['Average Wind'], RealTime_NP15['total'])
x_fit = np.linspace(1,14,500)
y_fit = [exp(x,*popt) for x in x_fit]
```

```python
p = figure(title='Daily Price vs Average Daily Wind Speed')
p.circle(x = 'Average Wind', y = 'total',source = RealTime_NP15)
p.line(x=x_fit,y=y_fit,line_width=3,line_color='red')
p.title.text_font_size='14pt'
p.yaxis.axis_label='Average Price'
p.yaxis.axis_label_text_font_size='14pt'
p.yaxis.major_label_text_font_size='12pt'
p.xaxis.axis_label='Average Wind Speed'
p.xaxis.axis_label_text_font_size='14pt'
p.xaxis.major_label_text_font_size='12pt'
p.legend.label_text_font_size='14pt'
#p.y_range.start = 20
p.xaxis.major_label_orientation = math.pi/4
show(p)
```

## Frequency Analysis

```python
Y = RealTime_NP15['total']
X = np.arange(len(Y))
```

```python
from scipy.stats import linregress
import scipy.fft
```

```python
slope, intercept, r_value, p_value, std_err = linregress(X,Y)
```

```python
Y_zeroed = Y-intercept-X*slope
array = np.pad(Y_zeroed,(0,10*len(Y_zeroed)),'constant', constant_values=(0, 0))
```

```python
fft = scipy.fft.fft(array)
```

```python
p = figure(title='Fourier Analysis')
p.line(x = range(1,len(fft)+1), y = list(np.abs(fft)))
p.title.text_font_size='14pt'
p.yaxis.axis_label='Average Price'
p.yaxis.axis_label_text_font_size='14pt'
p.yaxis.major_label_text_font_size='12pt'
p.xaxis.axis_label='Average Precipitation'
p.xaxis.axis_label_text_font_size='14pt'
p.xaxis.major_label_text_font_size='12pt'
p.legend.label_text_font_size='14pt'
#p.y_range.start = 20
p.xaxis.major_label_orientation = math.pi/4
show(p)
```
