```python
import requests
import simplejson as json
import numpy as np
import tabula
import pandas as pd
```

```python
def combine_on_off(row):
    if row['prop'] =='max':
        return max(row['ON'],row['OFF'])
    if row['prop'] =='mean':
        return row['ON']/4+3*row['OFF']/4
    return min(row['ON'],row['OFF'])
```

```python
from datetime import datetime
def parse_pdf(date):
    try:
        df = tabula.read_pdf('http://www.caiso.com/Documents/Real-TimeDailyMarketWatch'+date.strftime("%b%d_%Y")+'.pdf',pages='2',silent=True)[-1]
    except:
        try:
            df = tabula.read_pdf('http://www.caiso.com/Documents/Real-TimeDailyMarketWatch'+date.strftime("%b%d-%Y")+'.pdf',pages='2',silent=True)[-1]
        except:
            print(date.strftime("%b%d-%Y"))
    return df

def parse_pdf_day_ahead(date):
            
    url1 = 'http://www.caiso.com/Documents/Day-AheadDailyMarketWatch'+date.strftime("%b%d_%Y")+'.pdf'
    url2 = 'http://www.caiso.com/Documents/Day-AheadDailyMarketWatch'+date.strftime("%b%d-%Y")+'.pdf'
    try:
        #print(url1)
        df = tabula.read_pdf(url1,pages='1',silent=True)[-2]
    except:
        try:
            #print(url2)
            df = tabula.read_pdf(url2,pages='1',silent=True)[-2]
        except:
            print(date.strftime("%b%d-%Y"))
    return df

def transform_df(df,date):
    column_names=['status',('NP15','Max'),('NP15','Mean'),('NP15','Min'),('SP15','Max'),('SP15','Mean'),('SP15','Min'),
                  ('ZP26','Max'),('ZP26','Mean'),('ZP26','Min'),('PGAE','Max'),('PGAE','Mean'),('PGAE','Min'),
                  ('SCE','Max'),('SCE','Mean'),('SCE','Min'),('SDGE','Max'),('SDGE','Mean'),('SDGE','Min'),
                  ('VEA','Max'),('VEA','Mean'),('VEA','Min')]
    current_names = df.columns
    df = df.rename(columns=dict(zip(current_names,column_names)))[1:]
    df = df.set_index('status')
    dft = df.transpose()
    dft = dft.astype(float)
    regions = ['NP15','SP15','ZP26','PGAE','SCE','SDGE','VEA']
    region_array = [[region]*3 for region in regions]
    region_array = [val for sublist in region_array for val in sublist]
    Def_array = ['max','mean','min']*len(regions)
    dft['region']=region_array
    dft['prop'] = Def_array
    if 'ON' in dft.columns:
        dft['total'] = dft.apply (lambda row: combine_on_off(row), axis=1)
    else:
        dft['total']=dft['OFF']
    dft['date']=date
    dft = dft.set_index(['region','prop','date'])
    return dft
    
```

```python
dates = pd.date_range(start='1/2/2017', end='12/31/2018')
date1 = datetime.strptime('Jan01_2017',"%b%d_%Y")
dft = transform_df(parse_pdf(date1),date1)['total']
for date in dates:
    #print(date.strftime("%b%d_%Y"))
    try:
        dft = dft.append(transform_df(parse_pdf(date),date)['total'])
    except:
        pass
```

```python
dft.to_pickle('./realtime.pkl')
```
